```python
upper_list=[]
lower_list=[]

age_mean=dataSet.Age.mean()
age_std=dataSet.Age.std()
upper_list.append(age_mean+(age_std*3))
lower_list.append(age_mean-(age_std*3))
print('Age_mean:',age_mean,end=" ")
print('Age_std:',age_std)
bs_Fast_mean=dataSet.BS_Fast.mean()
bs_Fast_std=dataSet.BS_Fast.std()
upper_list.append(bs_Fast_mean+(bs_Fast_std*3))
lower_list.append(bs_Fast_mean-(bs_Fast_std*3))
print('bs_Fast_mean:',bs_Fast_mean,end=" ")
print('bs_Fast_std:',bs_Fast_std)
bs_pp_mean=dataSet.BS_pp.mean()
bs_pp_std=dataSet.BS_pp.std()
upper_list.append(bs_pp_mean+(bs_pp_std*3))
lower_list.append(bs_pp_mean-(bs_pp_std*3))
print('bs_pp_mean:',bs_pp_mean,end=" ")
print('bs_pp_std:',bs_pp_std)
plasma_R_mean=dataSet.Plasma_R.mean()
plasma_R_std=dataSet.Plasma_R.std()
upper_list.append(plasma_R_mean+(plasma_R_std*3))
lower_list.append(plasma_R_mean-(plasma_R_std*3))
print('plasma_R_mean:',plasma_R_mean,end=" ")
print('plasma_R_std:',plasma_R_std)
plasma_F_mean=dataSet.Plasma_F.mean()
plasma_F_std=dataSet.Plasma_F.std()
upper_list.append(plasma_F_mean+(plasma_F_std*3))
lower_list.append(plasma_F_mean-(plasma_F_std*3))
print('plasma_F_mean:',plasma_F_mean,end=" ")
print('plasma_F_std:',plasma_F_std)
hba1c_mean=dataSet.HbA1c.mean()
hba1c_std=dataSet.HbA1c.std()
print('hba1c_mean:',hba1c_mean,end=" ")
print('hba1c_std:',hba1c_std)
lower_list.append(hba1c_mean-(hba1c_std*3))
upper_list.append(hba1c_mean+(hba1c_std*3))


print('List=',upper_list)
print('List=',lower_list)
```

```python
X = dataSet.iloc[:,:-1].values
y = dataSet.iloc[:,6].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
from sklearn.preprocessing import StandardScaler
scale_X = StandardScaler()
X_train = scale_X.fit_transform(X_train)
X_test = scale_X.transform(X_test) 
```

### SVM

```python
from sklearn.svm import SVC
SVM_classifier = SVC(kernel = 'linear', gamma = 'auto', random_state = 0,probability=True)
SVM_classifier.fit(X_train, y_train)
SVM_y_pred = SVM_classifier.predict(X_test)
print(SVM_y_pred)
print(SVM_classifier.predict([[50,6.8,8.8,11.2,7.2,62]]))#type1
print(SVM_classifier.predict([[26,5.8,4.2,11.4,8.4,53]]))#type2
print(SVM_classifier.predict([[29,5.2,7.4,8.7,5.6,41]]))#normal
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_SVM = confusion_matrix(y_test, SVM_y_pred)
ac_SVM = accuracy_score(y_test, SVM_y_pred)
print("Confusion Matrix :")
print(cm_SVM)
print("SVM Train/test Split Accurary: ", ac_SVM)

pred_prob1 = SVM_classifier.predict_proba(X_test,'')
```

```python
from sklearn.metrics import roc_curve

# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_test, pred_prob1[:,1], pos_label=1)
#fpr2, tpr2, thresh2 = roc_curve(y_test, pred_prob2[:,1], pos_label=1)

# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```
